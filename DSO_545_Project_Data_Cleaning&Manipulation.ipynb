```python
df.dropna(inplace=True)  # This will remove rows with any missing values.
df
```

```python
new_esg_date = pd.to_datetime(df['ESG Score Date'], format='%m/%d/%Y').dt.strftime('%Y-%m-%d')
df['New ESG Date'] = new_esg_date
df.head()
tickers = df['Ticker']
results = []

# Loop through each ticker in the S&P 500 list
for ticker in tickers:
    try:
        # Fetch historical stock data
        stock_data = yfin.download(ticker, start='2018-01-01', end='2023-12-31')

        # Convert ESG score date to datetime (replace 'esg_score_date' with your ESG date column)
        filtered_df = df[df['Ticker'] == ticker]  # Replace with your ESG date column or specific ESG date
        target_date_str = filtered_df['New ESG Date'].values[0]

        esg_date = datetime.strptime(target_date_str, '%Y-%m-%d')

        # Check if adjusted close is available for the ESG score date
        if esg_date in stock_data.index:
            adj_close = stock_data.loc[esg_date, 'Adj Close']
            results.append({'Ticker': ticker, 'New ESG Date': target_date_str, 'Adjusted Close': adj_close})
        else:
            # Find the next trading day closest to the ESG score date
            next_trading_day = esg_date + timedelta(days=1)
            while next_trading_day not in stock_data.index:
                next_trading_day += timedelta(days=1)

            adj_close_next_day = stock_data.loc[next_trading_day, 'Adj Close']
            results.append({'Ticker': ticker, 'New ESG Date': target_date_str, 'Adjusted Close': adj_close_next_day})
    except Exception as e:
        print(f"Error processing {ticker}: {e}")

# Convert results to DataFrame
results_df = pd.DataFrame(results)
print(results_df)
```

```python
results_df = pd.DataFrame(results_df)
results_df.drop(columns=['New ESG Date','Ticker'], inplace=True)
df['Adjusted Close'] = results_df
df.head()
```

```python
yfin.pdr_override()
tickers1 = df['Ticker']
returns = []
for ticker in tickers1:
    if df.loc[df['Ticker'] == ticker, 'Year'].iloc[0] == 22:
      stock_data = pdr.get_data_yahoo(ticker, start='2022-01-01', end='2022-10-31')
    elif df.loc[df['Ticker'] == ticker, 'Year'].iloc[0] == 23:
      stock_data = pdr.get_data_yahoo(ticker, start='2023-01-01', end= '2023-10-31')
    monthly_returns = stock_data['Adj Close'].resample('M').ffill().pct_change()
    ytd_ret = monthly_returns.cumsum()
    ytd = ytd_ret.iloc[-1]
    returns.append({'Ticker': ticker, 'YTD Returns': ytd})
```

```python
yfin.pdr_override()
tickers2 = df['Ticker']
returns_before = []
for ticker in tickers2:
    if df.loc[df['Ticker'] == ticker, 'Year'].iloc[0] == 22:
      stock_data = pdr.get_data_yahoo(ticker, start='2021-03-01', end='2021-12-31')
    elif df.loc[df['Ticker'] == ticker, 'Year'].iloc[0] == 23:
      stock_data = pdr.get_data_yahoo(ticker, start='2022-03-01', end= '2022-12-31')
    monthly_returns = stock_data['Adj Close'].resample('M').ffill().pct_change()
    ytd_ret = monthly_returns.cumsum()
    ytd = ytd_ret.iloc[-1]
    returns_before.append({'Ticker': ticker, 'YTD Returns': ytd})
```
